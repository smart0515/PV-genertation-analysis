```python
pv = pv.rename(columns={
    '월': 'month',
    '일': 'day'
})

pv['datetime'] = pd.to_datetime(pv[['year', 'month', 'day']])

pv.drop(['month','day','year','sum'],axis=1,inplace=True)
```

```python
new_cols = {}
for col in pv.columns:
    if col.endswith('h'):
        new_cols[col] = col[:-1]
pv.rename(columns=new_cols, inplace=True)
```

```python
df=pv


# datetime 컬럼을 datetime 형태로 변환
df['datetime'] = pd.to_datetime(df['datetime'])

# 새로운 데이터프레임을 담을 리스트
new_data = []

# 각 행마다 반복하면서 시간 정보를 분리
for index, row in df.iterrows():
    date = row['datetime'].strftime('%Y-%m-%d')
    for hour in range(1, 25):
        time = f'{hour:02d}:00'
        new_data.append({
            'datetime': f'{date} {time}',
            'value': row[str(hour)]
        })

# 새로운 데이터프레임 생성
new_df = pd.DataFrame(new_data)
```

```python
pv = new_df
```

```python
plt.rc('font', family='Malgun Gothic')

selected_features = ['발전량', 'PM10', 'PM2.5', 'O3', '이산화질소','아황산가스','강수량(mm)','강수량 QC플래그','습도(%)','일조(hr)', '일조 QC플래그', '일사(MJ/m2)', '일사 QC플래그','전운량(10분위)','중하층운량(10분위)','시정(10m)','기온(°C)']
df_selected = df[selected_features]

corr = df_selected.corr()
plt.figure(figsize=(14, 10))
plt.title('Correlation Heatmap', fontsize=20)
sns.heatmap(corr, cmap='coolwarm', annot=True, vmin=-1, vmax=1)
```
