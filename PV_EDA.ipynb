```python
import os
import pandas as pd

# 각 파일의 경로를 지정합니다.
folder_path = "PVdata/"
files = os.listdir(folder_path)

# 모든 파일을 읽어서 하나의 데이터프레임으로 결합합니다.
dataframes = []
for file in files:
    if file.endswith(".csv"):
        df = pd.read_csv(os.path.join(folder_path, file),encoding='cp949')
        print(len(df))
        df["year"] = int(file[:4])  # 파일 이름에서 연도를 추출하여 year 열로 추가합니다.
        dataframes.append(df)

pv_all_data = pd.concat(dataframes, ignore_index=True)
```

```python
pv_all_data
```

```python
pv_all_data.info()
```

```python
pv_all_data.isnull().sum()
```

```python
year = pv_all_data['year']
```

```python
pv_all_data.drop('year',axis=1,inplace=True)
pv_all_data['sum']=pv_all_data.sum(axis=1)
```

```python
pv_all_data.head()
```

```python
pv_all_data['year']=year
```

```python
pv_all_data
```

```python
pv_all_data.to_csv('pv_data.csv',encoding='cp949')
```

## EDA

**전체기간 시간별 평균 발전량**

```python
# 각 column의 합계 계산
mean_by_hour = pv_all_data.mean(axis=0)[2:-2]

# 막대그래프 그리기
mean_by_hour.plot(kind='bar')
plt.title('mean by Hour')
plt.xlabel('Hour')
plt.ylabel('mean')
plt.show()
```

```python
# 연도별 전기 사용량 총합 구하기
yearly_sum = pv_all_data.groupby('year').sum()

# 막대그래프로 시각화
yearly_sum['sum'].plot(kind='bar')
plt.title('Yearly Electricity Consumption')
plt.xlabel('Year')
plt.ylabel('Electricity Consumption (kWh)')
plt.show()
```

**연도별 시간별 상대적인 차이**

```python
fig, axes = plt.subplots(6, 4, figsize=(20, 40))  # 24개 열을 3*8 그리드에 배치

for i, col in enumerate(pv_all_data.columns[2:-1]):  # 2열부터 23열까지 반복
    ax = axes[i//4][i%4]  # subplot 선택
    sns.barplot(x="year", y=col, data=pv_all_data, ax=ax)  # 막대 그래프 그리기
    ax.set_title(col)  # subplot 제목 설정

plt.rc('figure', titlesize=200)
plt.tight_layout(h_pad=2,w_pad=2)  # subplot 간격 조정
plt.show()  # 그래프 출력
```

**연도별 시간대별 상대적인 차이 - Ylim 고정**

```python
fig, axes = plt.subplots(3, 5, figsize=(20, 10))  # 24개 열을 3*8 그리드에 배치

for i, col in enumerate(pv_all_data.columns[7:-5]):  # 2열부터 23열까지 반복
    if i < 15:
        ax = axes[i//5][i%5]  # subplot 선택
    else:
        ax = axes[(i-17)//5][(i-15)%5]  # subplot 선택
    sns.barplot(x="year", y=col, data=pv_all_data, ax=ax)  # 막대 그래프 그리기
    ax.set_title(col)  # subplot 제목 설정
    ax.set_ylim([0, 1000])
    
plt.tight_layout(h_pad=3, w_pad=3)  # subplot 간격 조정

plt.show()  # 그래프 출력
```

```python
monthly_mean = pv_all_data.groupby('월').mean()
```

```python
import matplotlib.pyplot as plt

# 서브플랏 생성
fig, axs = plt.subplots(nrows=4, ncols=6, figsize=(20, 10))

# 각 시간대별로 반복하여 그래프 그리기
for i, ax in enumerate(axs.flatten()):
    hour = i + 1
    data = monthly_mean[f'{hour}h']
    data.plot(kind='bar', ax=ax)
    ax.set_title(f'Mean by Month ({hour}h)')
    ax.set_xlabel('Month')
    ax.set_ylabel('Mean')
    ax.set_ylim([0, 1000])
# 그래프 출력
plt.tight_layout()
plt.show()
```

# ASOS 데이터

```python
import os
import pandas as pd

# 각 파일의 경로를 지정합니다.
folder_path = "ASOS/"
files = os.listdir(folder_path)

# 모든 파일을 읽어서 하나의 데이터프레임으로 결합합니다.
dataframes = []
for file in files:
    if file.endswith(".csv"):
        df = pd.read_csv(os.path.join(folder_path, file),encoding='utf-8')
        print(len(df))
        df["year"] = int(file[:4])  # 파일 이름에서 연도를 추출하여 year 열로 추가합니다.
        dataframes.append(df)

asos_all_data = pd.concat(dataframes, ignore_index=True)
```

```python
missing_values = asos_all_data.isna().mean()
drop_columns = missing_values[missing_values > 0.5].index
asos_all_data.drop(drop_columns, axis=1, inplace=True)
```

```python
import seaborn as sns

selected_cols = ['기온(°C)', '기온 QC플래그', '풍속(m/s)', '풍속 QC플래그',
       '풍향(16방위)', '풍향 QC플래그', '습도(%)', '습도 QC플래그', '증기압(hPa)', '이슬점온도(°C)',
       '현지기압(hPa)', '현지기압 QC플래그', '해면기압(hPa)', '해면기압 QC플래그', '일조(hr)',
       '일조 QC플래그', '일사 QC플래그', '전운량(10분위)', '중하층운량(10분위)', '시정(10m)',
       '지면온도(°C)', '지면온도 QC플래그']
selected_df = asos_all_data[selected_cols]
corr = selected_df.corr()

plt.figure(figsize=(18, 14))
sns.heatmap(corr, annot=True, cmap='coolwarm')
```
